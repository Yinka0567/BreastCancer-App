Step 4: Data Cleaning

```python
df.drop(columns=["id"], inplace=True)
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
```

```python
# Count plot of diabetic vs. non-diabetic outcomes
sns.countplot(x='diagnosis', data=df, palette='Set2')
plt.title('Distribution of Breast Cancer Outcome')
plt.xlabel('Outcome (0 = No Breast Cancer, 1 = Breast Cancer)')
plt.ylabel('Count')
plt.show()
```

```python
# Plot distribution of all numeric features
df.hist(figsize=(15, 10), bins=20, color='skyblue', edgecolor='black')
plt.suptitle('Feature Distributions', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Box plots to observe distribution of features with respect to diabetes outcome
plt.figure(figsize=(16, 12))
for idx, col in enumerate(df.columns[:-1], 1):
    plt.subplot(3, 3, idx)
    sns.boxplot(x='diagnosis', y=col, data=df, palette='coolwarm')
    plt.title(f'{col} vs diagnosis')
    plt.tight_layout()
plt.show()
```

```python
# Compute correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
```

Step 5: Feature Scaling with StandardScaler Machine learning models often benefit from scaling. We'll use StandardScaler from sklearn.

```python
from sklearn.preprocessing import StandardScaler

# Separate features and target
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Step 6: Train-Test Split We split the data into training and testing sets using train_test_split.



```python
from sklearn.model_selection import train_test_split

# Split the scaled data into 80% training and 20% testing
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

```python
# Confirm split sizes
print("Training Set Shape:", X_train.shape)
print("X Testing Shape:", X_test.shape)
print("y Training Set Shape:", y_train.shape)
print("y Testing Shape:", y_test.shape)


```

Step 7: Training the model

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_scaled,y)
importances = rf.feature_importances_
features = pd.Series(importances, index = X.columns)
features.sort_values(ascending = False).plot(kind='barh', title = 'Random Forest Classifier')
plt.show()
print("Top Features by Random Forest:")
print(features.sort_values(ascending = False))
```

Step 8: Select 10 best features

```python
# Select top 10 features
top_10 = features.sort_values(ascending=False).head(10).index
X_top10 = X[top_10]

# Rescale top features
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(X_top10, y, test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Train Random Forest with top features
rf_top = RandomForestClassifier()
rf_top.fit(X_train_scaled, y_train)

```

```python
# Evaluate
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

y_pred = rf_top.predict(X_test_scaled)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
import joblib

joblib.dump(rf_top, "rf_top10_model.pkl")
joblib.dump(scaler, "scaler_top10.pkl")
joblib.dump(list(top_10), "top10_features.pkl")
```
